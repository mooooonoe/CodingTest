초기에 N개의 박스에 공이 각각 몇 개씩 들어 있었다.

이중 D 번째 박스에서 공을 모두 꺼내어 D+1 번째, D+2 번째, …, N번째, 1번째, … 이렇게 계속 반복하며 박스에 공을 하나씩 집어 넣었다.

제일 마지막에는 X번째 박스에 공을 집어넣고 더 이상 손에 남은 공이 없었다.

이 과정이 끝나고 나니 박스들에 공이 A[1], A[2], …, A[N]개가 들어있다.

A[1], A[2], ..., A[N]과, D, X가 입력으로 주어졌을 때, 초기에 D 번째 박스에 공이 몇 개 있었는지 구하라.

불가능한 경우는 주어지지 않는다.



아래 첫 번째 입력의 예시를 보자. 네 개의 박스에 각각 3, 2, 4, 5 개 공이 들어있고, D = 2, X = 3이다. 

입력에서 주어진 상자와 같은 상태로 만들기 위해선, 2 번 박스에 9개의 공이 필요하다. 

따라서 첫 번째 입력에 대한 출력 결과로 9를 출력한다. 





[제약사항]

1.   N은 4 이상 500 이하이다. (4 ≤ N ≤ 500)

2.   수열의 원소는 0 이상 20 이하의 정수이다.





[입력]

가장 첫 줄에는 테스트 케이스의 총 수가 주어진다.

그 다음 줄부터 각 테스트 케이스가 주어지며, 각 테스트 케이스는 2줄로 구성된다.

각 테스트 케이스의 첫 줄에는 N, D, X의 값이 주어진다.

다음 줄에 마지막 상태의 공 개수들이 주어진다.



[출력]

출력의 각 줄은 ‘#x’로 시작하고, 공백을 한 칸 두고 D번째 박스에 초기에 있던 공 개수를 출력한다.

단, x는 테스트 케이스의 번호이다.
